Objective

The objective of this notebook is to prepare the raw credit card transaction dataset for business analysis and dashboard development.

The data cleaning process includes handling unnecessary columns, converting date fields, and creating meaningful business features. Feature engineering is performed to generate customer demographics, transaction time attributes, and revenue-based customer segments.

The transformed dataset will be used for exploratory data analysis, business intelligence reporting, and Power BI dashboard development.

```python
import pandas as pd
```

```python
df = pd.read_csv("/content/credit_card_transactions.csv", nrows=100000)

df.shape
```

```python
# Remove useless column
df.drop('Unnamed: 0', axis=1, inplace=True)

# Convert dates
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

print(df.shape)
```

```python
df['age'] = 2026 - df['dob'].dt.year

df[['age']].head()
```

```python
bins = [18, 25, 35, 45, 55, 65, 100]

labels = ['18-25', '26-35', '36-45', '46-55', '56-65', '65+']

df['age_group'] = pd.cut(
    df['age'],
    bins=bins,
    labels=labels
)

df['age_group'].value_counts()
```

```python
df['year'] = df['trans_date_trans_time'].dt.year
df['month'] = df['trans_date_trans_time'].dt.month
df['day'] = df['trans_date_trans_time'].dt.day
df['hour'] = df['trans_date_trans_time'].dt.hour

df[['year', 'month', 'day', 'hour']].head()
```

```python
df['customer_name'] = df['first'] + ' ' + df['last']

df[['customer_name']].head()
```

```python
print(df.shape)
df.columns
```

```python
def revenue_segment(amount):
    if amount < 25:
        return "Low"
    elif amount < 75:
        return "Medium"
    elif amount < 200:
        return "High"
    else:
        return "Premium"

df['revenue_segment'] = df['amt'].apply(revenue_segment)

df['revenue_segment'].value_counts()
```

```python
print(df.shape)
```

```python
df.to_csv("fintech_cleaned.csv", index=False)

print("Dataset Exported Successfully")
```
